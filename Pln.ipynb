```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
texto = "Codar é muito legal e produtivo , viva a programação !"
tokenização_palavra = word_tokenize(texto)
tokenização_sentença=sent_tokenize(texto)
```
